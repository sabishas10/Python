```python
name=input("Enter your name:")
marks=int(input("Enter your marks:"))
if marks>100 or marks<0:
    grade="Invalid input"
elif marks>80:
    grade="A"
elif marks>=60:
    grade="B"
elif marks>=40:
    grade="C"
else:
    grade="Fail"
print("Name:",name)
print("Marks:",marks)
print("Grade:",grade)
```
