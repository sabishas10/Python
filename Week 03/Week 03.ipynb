## Exercise 3

```python
marks=int(input("Enter your marks:"))
if marks>=75:
    print("Grade A")
elif marks>=65:
    print("Grade B")
elif marks>=50:
    print("Grade C")
else:
    print("Grade F")
```

## Exercise 5

```python
num=int(input("Enter a number:"))
if num%2==0:
    print("Even number")
else:
    print("Odd number")
```

 ### OR

```python
num=int(input("Enter a number:"))
if num%2==1:
    print("Odd number")
else:
    print("Even number")
```

```python
name=input("Enter your name:")
Mark=int(input("Enter your marks:"))
if Mark>=75:
    grade="D"
elif Mark>=65:
    grade="C"
elif Mark>=50:
    grade="B"
else:
    grade="Fail"
print ("Name:",name,", Marks:",Mark,", Grade:",grade)
```
